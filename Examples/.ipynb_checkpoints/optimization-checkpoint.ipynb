# Splines

```python
# Define knots:
# Knot locations are at the centiles of the distribution of uncensored log event times,
# including all observed events, without omitting or filtering out any data points.
# - Boundary knots: placed at the 0th and 100th centiles (min and max values)
# - Internal knots: internal knots are placed at the centiles between the min and max   
knots = np.percentile(ln_t_train, [0, 25, 50, 75, 100])
knots
```

```python
# Initialize spline coefficients with seed for reproducibiliity 
np.random.seed(42)
gamma0 = np.random.uniform(size=(1, knots.size - 1))
gamma0 
```

```python
def spline_basis(x, k_j, k_min, k_max, derivative=False):
    """Computes the basis function S(x; k_j)."""
    phi_j = (k_max - k_j) / (k_max - k_min)
    
    def relu(x):
        return np.maximum(x, 0)  # (x - k)^3_+ is a ReLU-like function

    if derivative:
        # Derivative of the spline basis 
        return 3 * relu(x - k_j) ** 2 - 3 * phi_j * relu(x - k_min) ** 2 - 3 * (1 - phi_j) * relu(x - k_max) ** 2
    # Terms for the basis function 
    return relu(x - k_j) ** 3 - phi_j * relu(x - k_min) ** 3 - (1 - phi_j) * relu(x - k_max) ** 3


def spline_design_matrix(ln_t, knots):
    """Computes the spline function s(x; γ, k)."""
    # Boundary knots
    k_min, k_max = knots[0], knots[-1]
    # Construct basis functions over internal knots 
    basis = [spline_basis(ln_t, k_j, k_min, k_max) for k_j in knots[1:-1]]
    # Construct the design matrix 
    return np.array([ln_t] + basis)


D = []
for x in ln_t_train:
    # One spline equation per data point x
    D.append(spline_design_matrix(x, knots))
# Spline design matrix 
D = np.array(D)
D.shape 
```

```python
def spline_derivative_design_matrix(ln_t, knots):
    """Computes the spline function s(x; γ, k)."""
    # Boundary knots
    k_min, k_max = knots[0], knots[-1]
    # Construct basis functions over internal knots 
    basis = [spline_basis(ln_t, k_j, k_min, k_max, derivative=True) for k_j in knots[1:-1]]
    # Construct the design matrix 
    return np.array([1] + basis)
    

dDdt = []
for x in ln_t_train:
    # One spline equation per data point x
    dDdt.append(spline_derivative_design_matrix(x, knots))
# Design matrix for the spline derivative 
dDdt = np.array(dDdt)
dDdt.shape 
```

# Spline fitting using OLS 

```python
gamma_ols = (np.linalg.inv(D.T @ D) @ D.T @ ln_t_train)[None, :]
gamma_ols
```

```python
import matplotlib.pyplot as plt 
idx = np.argsort(ln_t_train)
plt.plot(ln_t_train[idx], marker="o", linestyle="", label="Data points")
plt.plot((D @ gamma_ols.T).squeeze()[idx], marker="o", linestyle="", label="Estimates")
plt.legend()

print("MSE of OLS spline fitting:", np.mean(((D @ gamma_ols.T).squeeze() - ln_t_train) ** 2))
```

# Spline fitting using gradient descent 

```python
gamma_var = tf.Variable(gamma0, dtype=tf.float32)

D_tf = tf.cast(D, dtype=tf.float32)
ln_t_train_tf = tf.cast(ln_t_train[:, None], dtype=tf.float32)


def loss():
    # Mean Squared Error 
    return tf.reduce_mean(tf.square(D_tf @ tf.transpose(gamma_var) - ln_t_train_tf))
    

optimizer = tf.keras.optimizers.SGD(learning_rate=0.005)

for epoch in tf.range(50):
    with tf.GradientTape() as tape:
        loss_value = loss()

    # Compute gradients
    gradients = tape.gradient(loss_value, [gamma_var])
    # Apply gradients to update gamma and beta
    optimizer.apply_gradients(zip(gradients, [gamma_var]))

    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Loss: {loss_value.numpy().mean()}")
```

```python
import numpy as np
import jax
import jax.numpy as jnp

from scipy.optimize import minimize

# Initialize spline coefficients with seed for reproducibility
np.random.seed(42)
gamma0 = np.random.uniform(size=knots.size - 1)
beta0 = np.random.uniform(size=X.shape[1])

# Convert initial variables to JAX arrays
gamma0 = jnp.array(gamma0, dtype=jnp.float64)
beta0 = jnp.array(beta0, dtype=jnp.float64)


# Define loss function using JAX
def loss_fn(vars_jnp):
    gamma, beta = jnp.split(vars_jnp, [gamma_dim])  
    eta = D @ gamma.T + X_train @ beta.T
    ln_ds_dx = jnp.log(jax.nn.softplus(dDdt @ gamma.T))  # Softplus ensures positivity   
    return -jnp.sum(e_train * (ln_ds_dx + eta) - jnp.exp(eta)).squeeze()

# Constraint function: ensure dDdt @ gamma > 0
def constraint_fn(vars_jnp):
    gamma, _ = jnp.split(vars_jnp, [gamma_dim])  
    return (dDdt @ gamma.T).squeeze()  # Must be > 0

# Define constraints for SciPy optimizer
constraint = {"type": "ineq", "fun": lambda vars_np: np.array(constraint_fn(jnp.array(vars_np)))}

# Initialize variables for optimization
gamma_dim = gamma0.size  
beta_dim = beta0.size
vars_init = jnp.concatenate([gamma0, beta0]) 

# Compute loss and gradient simultaneously
loss_and_grad_fn = jax.value_and_grad(loss_fn)

# Run SQP solver using SciPy
result = minimize(
    fun=lambda vars_np: loss_and_grad_fn(jnp.array(vars_np))[0],  # Loss function
    x0=np.array(vars_init),  # Ensure it's a NumPy array for SciPy
    jac=lambda vars_np: np.array(loss_and_grad_fn(jnp.array(vars_np))[1]),  # Gradient
    constraints=[constraint],
    method="SLSQP"
)

# Extract optimized gamma and beta
gamma_opt, beta_opt = jnp.split(jnp.array(result.x), [gamma_dim])
```

# De-centralized spline optimization using OLS 

```python
# Each pacticipant calculates 
# Z_i = D.T @ D
# y_i = D.T @ ln_t_train 
# Server calculates
# beta = sum_i(Z_i^{-1} @ y_hat_i)

# Datasets per participant 
n_participants = 4
data_split_idx = np.array_split(np.arange(X_train.shape[0]), n_participants)

Z_i, y_i = 0, 0 
for idx in data_split_idx:

    Z_i += D[idx].T @ D[idx]
    y_i += D[idx].T @ ln_t_train[idx]

gamma_ols_d = np.linalg.solve(Z_i, y_i)
gamma_ols_d
```

# De-centralized spline optimization using FedADMM 

```python
# GPT: 
# Can a leas squares problem be solved using the OLS method in the context of federated learning? 
# How can I solve for the regression coefficient using FedADMM

# Advantages:
# - Distributed & Privacy-Preserving: No raw data sharing.
# - Handles Non-IID Data: ADMM is more stable in federated settings.

rho = 1
penalty = rho * np.eye(D.shape[1])

lambdas = [np.zeros(D.shape[1]) for _ in range(n_participants)]  # Dual variables
betas = [np.zeros(D.shape[1]) for _ in range(n_participants)]     # Local models

beta_admm = np.zeros(D.shape[1])

for epoch in range(10):

    for i, idx in enumerate(data_split_idx):
        # Local update of each participant 

        A = D[idx].T @ D[idx] + penalty
        b = D[idx].T @ ln_t_train[idx] + rho * beta_admm - lambdas[i]
        betas[i] = np.linalg.solve(A, b)
    
    # Step 2: Global Update (Server)
    beta_new = np.mean(betas, axis=0)

    # Step 3: Dual Update (Clients)
    for i in range(n_participants):
        lambdas[i] += rho * (betas[i] - beta_new)

    # Convergence Check
    if np.linalg.norm(beta_new - beta_admm) < 1e-5:
        break

    beta_admm = beta_new
```

# Decentralized gradient descent  

```python
class Client:
    
    def __init__(self, beta, gamma):
        self.beta = beta
        self.gamma = gamma 

    def get_params(self):
        return self.beta, self.gamma 

    def set_params(self, beta, gamma):
        self.beta = beta 
        self.gamma = gamma 

    def fit(self, X, D, dD, t, e):
        
        def _loss():    
            eta = D @ tf.transpose(gamma) + X @ tf.transpose(beta)
            f1 = 1 / t * dD @ tf.transpose(gamma) * tf.math.exp(eta - tf.math.exp(eta))
            f2 = 1 / (1 + tf.math.exp(eta))        
            return -1 * tf.reduce_sum(tf.math.log(1 + e * f1 + (1 - e) * f2))
            
        beta = tf.Variable(self.beta, dtype=tf.float32, name="beta")
        gamma = tf.Variable(self.gamma, dtype=tf.float32, name="gamma")

        X = tf.cast(X, dtype=tf.float32)
        D = tf.cast(D, dtype=tf.float32)
        dD = tf.cast(dD, dtype=tf.float32)

        t = tf.cast(t, dtype=tf.float32)
        e = tf.cast(e, dtype=tf.float32)

        optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)
        for epoch in tf.range(5):

            with tf.GradientTape() as tape:
                loss_value = _loss()

            gradients = tape.gradient(loss_value, [gamma, beta])
            optimizer.apply_gradients(zip(gradients, [gamma, beta]))
    
        self.beta = beta.numpy()
        self.gamma = gamma.numpy()

        self.loss = float(loss_value.numpy().mean())
```

```python
# The gradient of the sum of log-likelihoods over all data is equal to the sum of the gradients of the log-likelihoods computed 
# over different splits of the data. This follows directly from the linearity of differentiation.
n_participants = 4

clients = [Client(beta_np, gamma_np) for _ in range(n_participants)]

idx = np.array_split(np.arange(X_train.shape[0]), n_participants)

for epoch in range(20):

    losses, betas, gammas = [], [], []
    for i, cli in enumerate(clients):
        cli.fit(X_train[idx[i]], D_train[idx[i]], dD_train[idx[i]], t_train[[idx[i]], None], e_train[[idx[i]], None])

        beta_i, gamma_i = cli.get_params()
        betas.append(beta_i)
        gammas.append(gamma_i)
        
        losses.append(cli.loss)

    beta_avg = sum(betas) / len(betas)
    gamma_avg = sum(gammas) / len(gammas)

    if epoch % 5 == 0:
        print(f"Epoch {epoch}, Loss: {np.mean(losses)}")
        
    for i, cli in enumerate(clients):
        cli.set_params(beta_avg, gamma_avg)
```

```python
beta_avg, gamma_avg = 0, 0
for i, cli in enumerate(clients):
    beta_i, gamma_i = cli.get_params()
    
    beta_avg += beta_i
    gamma_avg += gamma_i

beta_avg /= n_participants
gamma_avg /= n_participants
```

# Decentralized global aggregation

```python
class ClientGradient:
    
    def __init__(self, beta, gamma):
        self.beta = beta
        self.gamma = gamma 

    def get_params(self):
        return self.beta, self.gamma 

    def set_params(self, beta, gamma):
        self.beta = beta 
        self.gamma = gamma 

    def fit(self, X, D, dD, t, e):
        
        def _loss():    
            eta = D @ tf.transpose(gamma) + X @ tf.transpose(beta)
            f1 = 1 / t * dD @ tf.transpose(gamma) * tf.math.exp(eta - tf.math.exp(eta))
            f2 = 1 / (1 + tf.math.exp(eta))        
            return -1 * tf.reduce_sum(tf.math.log(1 + e * f1 + (1 - e) * f2))
            
        beta = tf.Variable(self.beta, dtype=tf.float32, name="beta")
        gamma = tf.Variable(self.gamma, dtype=tf.float32, name="gamma")

        X = tf.cast(X, dtype=tf.float32)
        D = tf.cast(D, dtype=tf.float32)
        dD = tf.cast(dD, dtype=tf.float32)

        t = tf.cast(t, dtype=tf.float32)
        e = tf.cast(e, dtype=tf.float32)

        with tf.GradientTape() as tape:
            loss_value = _loss()

        self.gradients = tape.gradient(loss_value, [gamma, beta])
        self.loss = float(loss_value.numpy().mean())
```

```python
beta_g = tf.Variable(beta_np, dtype=tf.float32)
gamma_g = tf.Variable(gamma_np, dtype=tf.float32)

clients = [ClientGradient(beta_np, gamma_np) for _ in range(n_participants)]

idx = np.array_split(np.arange(X_train.shape[0]), n_participants)

optimizer = tf.keras.optimizers.SGD(learning_rate=0.01)

for epoch in range(20):

    losses, gradients = [], []
    for i, cli in enumerate(clients):
        cli.fit(X_train[idx[i]], D_train[idx[i]], dD_train[idx[i]], t_train[[idx[i]], None], e_train[[idx[i]], None])
        
        gradients.append(cli.gradients)
        losses.append(cli.loss)
    
    grad_gamma, grad_beta = zip(*gradients)
    
    grad_gamma = tf.math.add_n(list(grad_gamma))
    grad_beta = tf.math.add_n(list(grad_beta))
    
    optimizer.apply_gradients(zip([grad_gamma, grad_beta], [gamma_g, beta_g]))
    
    for i, cli in enumerate(clients):
        cli.set_params(beta.numpy(), gamma.numpy())

beta.numpy(), gamma.numpy()
```

```python
def survival_curve(X, beta, t, gamma):

    S = [] 
    for k_j in knots[1:-1]:
        S.append(spline_basis(np.log(t), k_j))
    
    # Spline design matrix 
    D = np.array([1, np.log(t)] + S)[None, :]

    eta = (D @ gamma.T).squeeze() + (X @ beta.T).squeeze()
    return np.exp(-np.exp(eta))


# NOTE: Survival should be in [0, 1], decreasing over time 
survival_curve(X_train[0], beta.numpy(), 4, gamma.numpy())
```
